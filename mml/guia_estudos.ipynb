```python
# C√©lula 3: Fun√ß√£o Auxiliar para Atualizar Progresso

def completar_capitulo(df, chapter_number, confidence_level=90):
    """
    Marca um cap√≠tulo inteiro e suas se√ß√µes como 'Conclu√≠do' e define a data de estudo.
    """
    today = date.today()
    review_date = today + timedelta(days=30) # Agenda uma revis√£o para daqui a 30 dias
    
    # Seleciona todas as entradas do cap√≠tulo
    chapter_mask = df['Chapter'] == chapter_number
    
    # Atualiza as colunas
    df.loc[chapter_mask, 'Status'] = 'Conclu√≠do'
    df.loc[chapter_mask, 'StudyDate'] = today.strftime('%Y-%m-%d')
    df.loc[chapter_mask, 'ReviewDate'] = review_date.strftime('%Y-%m-%d')
    df.loc[chapter_mask, 'Confidence'] = confidence_level
    
    print(f"Cap√≠tulo {chapter_number} marcado como 'Conclu√≠do'!")
    return df
```

3.  **Adicione anota√ß√µes** na C√©lula 5 para se√ß√µes espec√≠ficas que voc√™ queira revisar.

5.  **Salve seu progresso** na C√©lula 7 para n√£o perder suas atualiza√ß√µes.

```python
# C√©lula 4: ATUALIZE SEU PROGRESSO AQUI

print("Marcando cap√≠tulos j√° conclu√≠dos...")
df = completar_capitulo(df, 2, confidence_level=70)
df = completar_capitulo(df, 3, confidence_level=60)
df = completar_capitulo(df, 4, confidence_level=50)
print("\n--- Progresso Atualizado! ---")

# Exibe o DataFrame com as linhas atualizadas
df[df['Status'] == 'Conclu√≠do']
```
